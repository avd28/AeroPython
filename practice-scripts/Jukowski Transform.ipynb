```python
from matplotlib import pyplot
%matplotlib inline
```

```python
def add_style(title, x_label, y_label):
    fig = pyplot.figure()
    pyplot.grid(True)
    pyplot.xlabel(x_label, fontsize=16)
    pyplot.ylabel(y_label, fontsize=16)
    pyplot.title(title, fontsize=16);
        
    return fig
```

```python
def plot_complex(e):
    real = e.real
    imag = e.imag
    fig = add_style('', 'x' , 'y')
    pyplot.plot(real,imag,lw=2, ls='--');
    pyplot.show()

plot_complex(e)
```

```python
del_x = 0.1
del_y = 0.1
c = 1
radius = np.sqrt((c-del_x)**2 + del_y**2)
z = complex_circle(radius)
real_z = z.real - del_x
imag_z = z.imag + del_y
z_new = real_z + (1j)*imag_z
e = jukowski(z_new,1)
plot_complex(e)
```
